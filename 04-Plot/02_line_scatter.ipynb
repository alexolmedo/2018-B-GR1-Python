```python
import numpy as np
import matplotlib.pyplot as plt
# https://stackoverflow.com/questions/43027980/purpose-of-matplotlib-inline/43028034
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Basic plot seno
xvals = np.arange(0,10,0.1)
plt.plot(xvals, np.sin(xvals))
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
# Tipos de imagenes para el simple plot
# Solido
# Guiones 
# Puntos
# Guion Punto
idx = 0
for marker in ('-', '--', ':', '-.'):
    plt.plot(xvals, np.sin(xvals)+idx, ''.join(('k',marker)))
    idx += 1
```

```python
# Marcadores
plt.plot(xvals, np.sin(xvals), 'k.--')  # .-- Se puede combinar
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Marcadores frecuencia 'markevery'
xvals2 = np.arange(0,10,0.01)
plt.plot(xvals2, np.sin(xvals2), 'k.--', markevery=50 ) # , markevery=10 
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```

```python
# Marcadores tamano y color
plt.plot(xvals, np.sin(xvals), 'r.', markersize=1, markevery=3) # 30 40   # r = red k = black b = blue etc...
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
```
