```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)), sigma=10)
```

```python
# Basico
plt.imshow(rands2d)
plt.colorbar()
```

```python
# Limites del grafico
plt.imshow(rands2d, extent=[-200,100,-100,100])
```

```python
# Minimo y maximo valor de la escala
plt.imshow(rands2d, vmax=0.3, vmin= -0.05) 
plt.colorbar()
```

```python
# Aspect ratio
plt.imshow(rands2d, aspect=0.2) # 0.... 1 horizontal, vertical
```

```python
# Interpolacion
small = np.arange(0,4,1).reshape(2,2)
print(small)
plt.imshow(small, interpolation='none')

plt.matshow(small)  # helper function, deshabilita la interpolacion y usa los valores en el centro
```

```python
# CONTORNOS

# Basico
srands2d = gaussian_filter(rands2d, sigma=100)
plt.imshow(srands2d)
plt.colorbar()
```

```python
# Contorno
plt.contour(srands2d)
plt.colorbar()
```

```python
# Niveles (3D)
plt.contour(srands2d, levels=[0,0.003, 0.045])
```

```python
# Numero de contornos
plt.contour(srands2d, 30)
```

```python
# Colores del contorno
plt.contour(srands2d, 30, colors=['r','b'])
```

```python
# Contornos llenos
plt.contourf(srands2d, 10)
```

```python
# Rejillas
plt.contourf(srands2d, 6, hatches=['*','-','.'])
```

```python
# Estilo de la linea
plt.contour(srands2d, linestyles=[':','--','-'])
```

```python
# labels
contours = plt.contour(srands2d, 5)
plt.clabel(contours)
```

```python
# Multiple
plt.contour(srands2d, colors=['k']) # levels=[0],
plt.imshow(srands2d)
```
